### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#total number of players (unique values, no duplicates)
players = purchase_data["SN"].nunique()
total_players = pd.DataFrame({"Total Players": [players]}, columns = ["Total Players"])
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#number of unique items 
unique_items = len(purchase_data["Item ID"].value_counts())
unique_items
```

```python
#average price
average_price = purchase_data["Price"].mean()
average_price
```

```python
#number of purchases
purchases = purchase_data["Price"].count()
purchases
```

```python
#total revenue
total_revenue = purchase_data["Price"].sum()
total_revenue
```

```python
#create summary data frame
summary_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                          "Average Price": [average_price],
                          "Number of Purchases": [purchases],
                          "Total Revenue": [total_revenue]})
summary_df["Average Price"] = summary_df["Average Price"].map('${:.2f}'.format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map('${:.2f}'.format)
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#I set the gender as the index 
gender_group = purchase_data.set_index('Gender')
gender_group.head()
```

```python
#groupby gender/ unique values 
unique_gender = gender_group.groupby('Gender').nunique()["SN"].rename("Total Count")
gender_df = pd.DataFrame(unique_gender)
gender_df
```

```python
#percentage and count per gender 
gender_df["Percentage of Players"] = gender_df/players * 100
gender_df
```

```python
#sort values
gender_df = gender_df.sort_values("Total Count", ascending=False)
#format values
gender_df["Percentage of Players"] = gender_df["Percentage of Players"].map('{:.2f}%'.format)
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#purchase count by gender 
purchase_count = gender_group.groupby('Gender').count()["Item ID"].rename("Purchase Count")
purchase_count =pd.DataFrame(purchase_count)
purchase_count
```

```python
#average purchase price
ave_pp_gender = gender_group.groupby('Gender').mean()["Price"].rename("Average Purchase Price")
ave_pp_gender = pd.DataFrame(ave_pp_gender)
ave_pp_gender
```

```python
#total purchase value 
total_value = gender_group.groupby('Gender').sum()["Price"].rename("Total Purchase Value")
total_value = pd.DataFrame(total_value)
total_value
```

```python
#creating new df by merging df's 
analysis_gender = purchase_count.merge(ave_pp_gender, on= "Gender")
analysis_gender = analysis_gender.merge(total_value, on="Gender")
analysis_gender
```

```python
#average total per person
analysis_gender["Avg Total Purchase per Person"] = total_value["Total Purchase Value"]/gender_df["Total Count"]
analysis_gender
```

```python
#formatting 
analysis_gender = pd.DataFrame(analysis_gender)
analysis_gender.style.format({"Average Purchase Price": '${:.2f}',
                             "Total Purchase Value": '${:.2f}',
                             "Avg Total Purchase per Person": '${:.2f}' })
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#establish bins (age)
age_bin = [0, 9, 14, 19, 24, 29, 34, 39, 99]
bin_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
#categorize existing players
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], age_bin, labels = bin_labels)
purchase_data.head()
```

```python
# create group based off the bin / set index 
grouped_age = purchase_data.set_index("Age Group")
grouped_age = pd.DataFrame(grouped_age)
grouped_age.head()
```

```python
age_demo = grouped_age.groupby("Age Group").nunique()["SN"].rename("Total Count")
age_demo = pd.DataFrame(age_demo)
age_demo
```

```python
#calculate percentages 
age_demo["Percentage of Players"] = age_demo/players *100
age_demo.index.name =None
age_demo = pd.DataFrame(age_demo)
#formatting 
age_demo.style.format({"Percentage of Players":"{:,.2f}%"})
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#purchase count
pcount_age = grouped_age.groupby('Age Group').nunique()["Purchase ID"].rename("Purchase Count")
pcount_age = pd.DataFrame(pcount_age)
pcount_age
```

```python
#average purchase price by age
ave_pp_age = grouped_age.groupby('Age Group').mean()["Price"].rename("Average Purchase Price")
ave_pp_age = pd.DataFrame(ave_pp_age)
ave_pp_age
```

```python
#total purchase value 
tpv_age = grouped_age.groupby('Age Group').sum()["Price"].rename("Total Purchase Value")
tpv_age = pd.DataFrame(tpv_age)
tpv_age
```

```python
#merging df's
pa_age = pcount_age.merge(ave_pp_age, on="Age Group")
pa_age = pa_age.merge(tpv_age, on="Age Group")
pa_age
```

```python
#average total purchase per person
pa_age["Avg Total Purchase per Person"] = tpv_age["Total Purchase Value"]/age_demo["Total Count"]
pa_age
```

```python
#formatting
pa_age.index.name = None
pa_age = pd.DataFrame(pa_age)
pa_age.style.format({"Average Purchase Price": '${:.2f}',
                     "Total Purchase Value": '${:.2f}',
                     "Avg Total Purchase per Person": '${:.2f}' })
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#set SN as index
sn = purchase_data.set_index("SN")

#purchase count by SN
sn_count = sn.groupby('SN').nunique()["Purchase ID"].rename("Purchase Count")
purchase_sn = pd.DataFrame(sn_count)
purchase_sn.head()
```

```python
#average purchase price per sn
app_sn = sn.groupby("SN").mean()["Price"].rename("Average Purchase Price")
app_sn = pd.DataFrame(app_sn)
app_sn.head()
```

```python
#total purchase value 
total_value_sn = sn.groupby('SN').sum()["Price"].rename("Total Purchase Value")
total_value_sn = pd.DataFrame(total_value_sn)
total_value_sn.head()
```

```python
top_spenders = purchase_sn.merge(app_sn, on="SN")
top_spenders = top_spenders.merge(total_value_sn, on="SN")
top_spenders.head()
```

```python
#formatting & sorting
top_spenders = pd.DataFrame(top_spenders)
top_spenders_formatted = top_spenders.sort_values('Total Purchase Value', ascending=False).head()
top_spenders_formatted.style.format({"Average Purchase Price":'${:.2f}',"Total Purchase Value":'${:.2f}'})
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#groupby item ID and item Name 
# obtain purchase count
most_popular = purchase_data.groupby(["Item ID","Item Name"]).count()["Purchase ID"].rename("Purchase Count")
most_popular = pd.DataFrame(most_popular)
most_popular.head()
```

```python
#item price
item_price = purchase_data.groupby(["Item ID","Item Name"]).mean()["Price"].rename("Items Price")
item_price = pd.DataFrame(item_price)
item_price.head()
```

```python
#total purchase value 
tp_value = purchase_data.groupby(["Item ID","Item Name"]).sum()["Price"].rename("Total Purchase Value")
tp_value = pd.DataFrame(tp_value)
tp_value.head()
```

```python
most_pop_item = most_popular.merge(item_price, on=["Item ID","Item Name"])
most_pop_item = most_pop_item.merge(tp_value, on =["Item ID","Item Name"])
most_pop_item = pd.DataFrame(most_pop_item)
most_pop_item.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#sorting/ formatting
most_pop_sorted = most_pop_item.sort_values("Purchase Count", ascending= False).head()
most_pop_sorted.style.format({"Items Price": '${:.2f}', "Total Purchase Value": '${:.2f}'})
```
